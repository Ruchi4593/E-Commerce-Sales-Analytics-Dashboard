```python
customer_data = df.groupby('Customer ID').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique'
}).reset_index()

customer_data.rename(columns={
    'Order ID': 'Total_Orders'
}, inplace=True)

customer_data.head()
```

```python
customer_data.shape
```

```python
customer_data.describe()
```

```python
features = customer_data[['Sales', 'Profit', 'Total_Orders']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_features = scaler.fit_transform(features)
```

```python
scaled_features[:5]
```

```python
from sklearn.cluster import KMeans

wcss = []

for k in range(1, 11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    
    kmeans.fit(scaled_features)
    
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker='o')

plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')

plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

customer_data['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
customer_data.head()
```

```python
customer_data['Cluster'].value_counts()
```

```python
cluster_summary = customer_data.groupby('Cluster').agg({
    'Sales': 'mean',
    'Profit': 'mean',
    'Total_Orders': 'mean'
}).round(2)

cluster_summary
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=customer_data,
    x='Sales',
    y='Profit',
    hue='Cluster'
)

plt.title('Customer Segments')
plt.show()
```

```python
cluster_names = {
    0: 'Medium Value',
    1: 'Low Value',
    2: 'High Value'
}

customer_data['Customer Segment'] = customer_data['Cluster'].map(cluster_names)
```
