# E-Commerce Sales Analytics Dashboard

## Project Objective

The objective of this project is to analyze e-commerce sales data to identify business trends, profitability drivers, customer behavior, and high-value customer segments.

The project combines Data Analytics and Machine Learning techniques to support business decision-making.

# Dataset Overview

Dataset: Sample Superstore Dataset

Total Records: 9994

Total Columns: 21

Key Fields:
- Sales
- Profit
- Quantity
- Discount
- Category
- Sub-Category
- Region
- State
- Customer Name
- Order Date

# Data Cleaning

Performed the following steps:

- Loaded dataset using Pandas
- Resolved encoding issues using Latin-1 encoding
- Converted Order Date and Ship Date to datetime format
- Created Year and Month features
- Checked missing values
- Checked duplicate records

Results:
- Missing Values: 0
- Duplicate Records: 0

# Business Insights

1. Technology is the highest revenue-generating category.

2. Technology is also the most profitable category.

3. Furniture generates high sales but low profit.

4. Tables are the largest loss-making sub-category.

5. Higher discounts are associated with lower profitability.

6. West region generates the highest sales and profit.

7. California is the top-performing state.

8. Texas is the largest loss-making state.

9. Sales increase significantly from September to December.

10. November is the highest sales month.
